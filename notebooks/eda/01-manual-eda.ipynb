```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv("../data/train.csv")
```

podemos ver que a quantidade de dados que temos para treinamento é muito pequena, então temos que nos atentar ao feature engineering e a escolha do modelo para que possamos ter um bom resultado.

```python
# get the outliers
for col in data.columns[:-1]:
        if data[col].dtype.kind in 'iufc':
            iqr = data[col].quantile(0.75) - data[col].quantile(0.25)
            lower_bound = data[col].quantile(0.25) - 1.5 * iqr
            upper_bound = data[col].quantile(0.75) + 1.5 * iqr

            count_outliers = len(data[(data[col] < lower_bound) | (data[col] > upper_bound)])

            if count_outliers > 0:
                print(f"{col}: {count_outliers}")
```

```python
sns.set_theme(style='whitegrid')

# Plotar a distribuição da receita
plt.figure(figsize=(10, 6))
sns.histplot(data['revenue'], kde=True, color='blue')
plt.title('Distribuição da Receita')
plt.xlabel('Receita')
plt.ylabel('Frequência')
plt.show()

# plotar o boxplot da receita
plt.figure(figsize=(10, 6))
sns.boxplot(data['revenue'], color='blue')
plt.title('Boxplot da Receita')
plt.xlabel('Receita')
plt.show()
```

podemos ver que a distribuição de receita não possui uma distribuição normal, o que é um problema para a maioria dos modelos de machine learning. Para resolver isso, podemos aplicar uma transformação logarítmica na variável alvo e, em seguida, verificar a distribuição da variável alvo após a transformação.

```python
# Plotar a distribuição da receita aplicado o logaritmo
plt.figure(figsize=(10, 6))
sns.histplot(np.log1p(data['revenue']), kde=True, color='blue')
plt.title('Distribuição da Receita (log)')
plt.xlabel('Receita')
plt.ylabel('Frequência')
plt.show()

# # pra voltar ao normal, reverter o log, aplicando o exp
# plt.figure(figsize=(10, 6))
# sns.histplot(np.expm1(np.log1p(data['revenue'])), kde=True, color='blue')
# plt.title('Distribuição da Receita (exp)')
# plt.xlabel('Receita')
# plt.ylabel('Frequência')
# plt.show()

# plotar o boxplot da receita aplicando o logaritmo
plt.figure(figsize=(10, 6))
sns.boxplot(np.log1p(data['revenue']), color='blue')
plt.title('Boxplot da Receita (log)')
plt.xlabel('Receita')
plt.show()
```

```python
# Analisar correlação entre as variáveis numéricas
corr = data.drop(columns="Id").select_dtypes(include='number').corr(method="pearson")

# Exibir a matriz de correlação
plt.figure(figsize=(16, 12))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, annot_kws={"size": 8})
plt.title('Matriz de Correlação')
plt.show()
```

```python
threshold = 0.97
corr_pairs = corr.unstack().sort_values(kind="quicksort").drop_duplicates()
high_corr_pairs = corr_pairs[(corr_pairs > threshold) & (corr_pairs < 1)]
print(high_corr_pairs)
```

```python
corr_with_revenue = corr['revenue'].sort_values(ascending=False)
print(corr_with_revenue)
```

```python
# Receita média por City (apenas as 10 maiores cidades por receita média)
top_cities = data.groupby('City')['revenue'].mean().sort_values(ascending=False).head(15).index
plt.figure(figsize=(12, 6))
sns.boxplot(x='City', y='revenue', data=data[data['City'].isin(top_cities)])
plt.title('Receita por City (Top 15)')
plt.xlabel('City')
plt.ylabel('Receita')
plt.xticks(rotation=45)
plt.show()
```

Ao olharmos somente para as 15 top cidades já podemos perceber que há uma diferença significativa do revenue entre elas. Algumas cidades como Izmir, Instanbul, Bursa e Kayseri possuem um revenue evidentemente maior que as outras cidades.

```python
# a data de abertura tem alguma relação com a receita?
data['Open Date'] = pd.to_datetime(data['Open Date'], format='%m/%d/%Y')

data['Open Year'] = data['Open Date'].dt.year

plt.figure(figsize=(10, 6))
sns.boxplot(x='Open Year', y='revenue', data=data)
plt.title('Receita por Ano de Abertura')
plt.xlabel('Ano de Abertura')
plt.ylabel('Receita')
plt.show()
```

Restaurantes abertos há muito tempo não obrigatoriamente tem receita maior, porém quando abertos nos anos 2000, parece ter um destaque na receita. Os demais anos parecem estáveis em relação à receita, mudando apenas em 2013 e 2014, com uma receita mediana mais baixa que as demais. Isso mostra que a variável de ano pode ser uma boa variável para o modelo.

```python
data["OpenedDays"] = (data['Open Date'].max() - data['Open Date']).dt.days

plt.figure(figsize=(10, 6))
sns.scatterplot(x='OpenedDays', y='revenue', data=data)
plt.title('Receita por Dias Abertos')
plt.xlabel('Dias Abertos')
plt.ylabel('Receita')
plt.show()
```

Aparentemente não existe uma relação direta, mas conseguimos considerar como uma variável para o modelo.

```python
numeric_features = data.select_dtypes(include='number').columns

fig, axes = plt.subplots((len(numeric_features) - 1) // 3 + 1, 3, figsize=(18, 30))

for i, feature in enumerate(numeric_features):
    ax = axes[i // 3, i % 3]
    sns.histplot(data[feature], kde=True, ax=ax)
    plt.ylim(0)

plt.tight_layout()
```
